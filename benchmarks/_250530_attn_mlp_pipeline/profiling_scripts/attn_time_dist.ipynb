```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

total_samples = len(attn_time_dist_32k)
# Create histogram
fig = plt.figure(figsize=(10, 6))
ax = plt.gca()
n, bins, patches = plt.hist(attn_time_dist_32k, bins=20, edgecolor='black')

# Add value annotations for each bar
for i, patch in enumerate(patches):
    height = n[i]
    x = patch.get_x() + patch.get_width()/2
    y = height
    ax.text(x, y, f'{int(height)}', ha='center', va='bottom')

# Add tooltips
def hover(event):
    for i, patch in enumerate(patches):
        if patch.contains(event)[0]:
            # Get bin info
            bin_start = bins[i]
            bin_end = bins[i+1]
            count = n[i]
            # Show tooltip
            tooltip = f'Range: {bin_start:.1f}-{bin_end:.1f}ms\nCount: {int(count)}'
            ax.annotate(tooltip,
                       xy=(event.xdata, event.ydata),
                       xytext=(10, 10), textcoords='offset points',
                       bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                       arrowprops=dict(arrowstyle='->'))
            fig.canvas.draw_idle()
            return
    # Remove tooltip if not hovering over a bar
    ax.texts.clear()
    fig.canvas.draw_idle()

plt.xlabel('Time (ms)')
plt.ylabel('Frequency') 
plt.title(f'Distribution of Execution Times - 32K Attention Sampling (total samples = {total_samples})')

# Add mean and std lines
mean = np.mean(attn_time_dist_32k)
std = np.std(attn_time_dist_32k)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}ms')
plt.axvline(mean + std, color='green', linestyle='dashed', linewidth=1, label=f'Mean  Std: {std:.2f}ms')
plt.axvline(mean - std, color='green', linestyle='dashed', linewidth=1)

plt.legend()
plt.grid(True, alpha=0.3)

fig.canvas.mpl_connect('motion_notify_event', hover)
plt.show()
```
